##  Tensors
- Each tensors has a data type and shape
- Mostly will be using numbers but there are strings and other too
- Shape = representation of the tensor 
- `tf.Variable(value, datatype)`
- Scalar = 0 degree (rank), but there are other tensors with higher degrees/rank such as matrix/vectors where there are more dimensions

````python
string = tf.Variable("this is a string", tf.string)
number = tf.Variable(123, tf.int16)
floating = tf.Variable(3.1415, tf.float64)
rank1_tensors = tf.Variable(["Hello", "Bob"], tf.string)
rank2_tensors = tf.Variable([["John", "Bob", "Frank"], ["Joe", "Helen", "Julie"]], tf.string) #matrix
````

You can use `tf.rank` to determine what rank the Tensor is

````python
tf.rank(string)
tf.rank(rank1_tensors)
tf.rank(rank2_tensors)
````

`.shape` - how many elements in each list [length of list, how many in each list]

````python
rank2_tensors.shape
````

## Reshape Tensor
- You can also change the shape of the tensors
- eg. You can change a rank 2 tensor into a rank 1
- `tf.ones()` make a list of ones
- `tf.reshape(tensor, shape)` to reshape the tensor

````python
tensor1 = tf.ones([1,2,3]) 
tensor2 = tf.reshape(tensor1, [2,3,1]) #reshape exsisting tensor to a new shape
tensor3 = tf.reshape(tensor2, [3, -1]) # -1 infer what the number need to be, it will calculate it for us
````

However, the number of element in the reshape should be the same as the original

````python
print(tensor1)
print(tensor2)
print(tensor3)
````
